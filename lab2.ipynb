```python
#15Probability density distribution of continuous numerical variable-length
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Car_sales.csv")

# Clean the 'Length' column (convert to numeric, handle non-numeric values like '.')
df['Length'] = pd.to_numeric(df['Length'], errors='coerce')  # Converts invalid values to NaN

# Drop missing values in 'Length'
df_clean = df.dropna(subset=['Length'])

# Plot the probability density distribution
sns.kdeplot(data=df_clean, x='Length', fill=True, color='skyblue', linewidth=2)

# Add labels and title
plt.title('Probability Density Distribution of Car Length')
plt.xlabel('Length')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
#27 Apply feature scaling on numerical variables
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load dataset
df = pd.read_csv("Car_sales.csv")

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Initialize scalers
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# Apply StandardScaler (zero mean, unit variance)
df_standardized = df.copy()
df_standardized[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols])

# Apply MinMaxScaler (scales between 0 and 1)
df_normalized = df.copy()
df_normalized[numerical_cols] = minmax_scaler.fit_transform(df[numerical_cols])

# Display results
print("Standardized Data:\n", df_standardized.head())
print("\nNormalized Data:\n", df_normalized.head())


```
